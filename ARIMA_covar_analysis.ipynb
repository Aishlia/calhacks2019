```python
import pandas as pd
import numpy as np
from pandas import datetime
from pandas import read_csv
from pandas import datetime
from pandas import DataFrame
from matplotlib import pyplot
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("SP500.csv", index_col=0, usecols=["Date", "Price"])
data['Price'] = data['Price'].str.replace(',','').astype(float)
```

An ARIMA model, or anything similar, will require three variable: p, q, r. To estimate the value of these variables, we will analyze our dataset using the following methods. 

```python
# Original Series
fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(data.Price); axes[0, 0].set_title('Original Series')
plot_acf(data.Price, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(data.Price.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(data.Price.diff().dropna(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(data.Price.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(data.Price.diff().diff().dropna(), ax=axes[2, 1])

plt.show()
```

Even the first order differencing looks almost over differenced. Therefore, we will attempt to use 1st order and 0th order. 1st order differencing achieves acceptable stationarity.

```python
from pmdarima.arima.utils import ndiffs

# Augmented Dickey-Fuller test
print(ndiffs(data.Price, test='adf'))

# Kwiatkowski–Phillips–Schmidt–Shin test
print(ndiffs(data.Price, test='kpss'))

# Phillips–Perron test:
print(ndiffs(data.Price, test='pp'))
```

Based on the results of these tests, we will use 1st order differencing (d=1). 

```python
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(data.Price.diff()); axes[0].set_title('1st Differencing')
axes[1].set(ylim=(0,5))
plot_pacf(data.Price.diff().dropna(), ax=axes[1])

plt.show()
```

Displayed are the ACF and PACF plots. We will use an autoregressive order of 1 (p=1).

```python
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

# Import data
fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(data.Price.diff()); axes[0].set_title('1st Differencing')
axes[1].set(ylim=(0,1.2))
plot_acf(data.Price.diff().dropna(), ax=axes[1])

plt.show()
```

We will use a moving average order of 1 (q=1).

Parse the data to convert the date information into pandas datetime and create a pandas dataframe containing all of the data. 

```python
def parser(x):
    return datetime.strptime(x, '%b %d, %Y')

data = pd.read_csv("SP500.csv", index_col=0, usecols=["Date", "Price"], parse_dates=[0], date_parser=parser)
data['Price'] = data['Price'].str.replace(',','').astype(float)
data = data.reindex(index=data.index[::-1])
```

Create an ARIMA model with terms obtained earlier:
p=3 (order of the AR term),
d=1 (number of differencing to make the time series stationary), and 
q=1 (order of the MA term).

```python
model = ARIMA(data, order=(3,1,1))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

```python
residuals = DataFrame(model_fit.resid)
residuals.plot()
pyplot.show()
```

```python
residuals.plot(kind='kde')
pyplot.show()
```

```python
print(residuals.describe())
```

```python
dataset = data.values
size = int(len(dataset) * 0.90)
train, test = dataset[0:size], dataset[size:len(dataset)]
history = [x for x in train]
predictions = list()
```

```python
for t in range(len(test)):
	model = ARIMA(history, order=(3,1,1))
	model_fit = model.fit(disp=0)
	output = model_fit.forecast()
	yhat = output[0]
	predictions.append(yhat)
	obs = test[t]
	history.append(obs)
	# print('predicted=%f, expected=%f' % (yhat, obs))
```

```python
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
```
